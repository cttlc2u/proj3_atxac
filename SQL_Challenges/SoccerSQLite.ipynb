```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('database.sqlite')
```

```python
query = "SELECT * FROM sqlite_master"

df_schema = pd.read_sql_query(query, conn)

df_schema.tbl_name.unique()
```

```python
query1 = 'SELECT home_team_api_id, sum(home_team_goal) AS total_home_goals FROM "Match" GROUP BY home_team_api_id ORDER BY sum(home_team_goal) DESC;'
pd.read_sql_query(query1, conn)
```

```python
query2 = 'SELECT away_team_api_id, sum(away_team_goal) AS total_away_goals FROM "Match" GROUP BY away_team_api_id ORDER BY sum(away_team_goal) DESC;'
pd.read_sql_query(query2, conn)
```

```python
query3 = 'SELECT count(*) AS total_tied_games FROM "Match" WHERE home_team_goal = away_team_goal;'
pd.read_sql_query(query3, conn)
```

```python
query4a = 'SELECT count(*) as mr_smiths FROM "Player" WHERE player_name LIKE "%Smith";'
pd.read_sql_query(query4a, conn)
```

```python
query4b = 'SELECT count(*) as has_smith FROM "Player" WHERE player_name LIKE "%smith%";'
pd.read_sql_query(query4b, conn)
```

```python
query4c = 'SELECT * FROM "Player" WHERE player_name LIKE "%smith%";'
pd.read_sql_query(query4c, conn)
```

```python
query5 = 'WITH tied_games AS (SELECT * FROM "Match" WHERE home_team_goal = away_team_goal ORDER BY home_team_goal ASC LIMIT (6596/2)+1), \
middle_values AS (SELECT home_team_goal, away_team_goal FROM "tied_games" ORDER BY home_team_goal DESC LIMIT 2) \
SELECT sum(home_team_goal)/2 as median_tie_score FROM "middle_values";'
pd.read_sql_query(query5, conn)
```

```python
query6 = 'WITH pref_foot AS (SELECT preferred_foot,count(preferred_foot) AS lefties FROM "Player_Attributes" WHERE preferred_foot="left") \
SELECT pref_foot.preferred_foot, (lefties*100)/count(Player_Attributes.preferred_foot) AS percent_left_footed FROM "pref_foot","Player_Attributes";'
pd.read_sql_query(query6, conn)
```
